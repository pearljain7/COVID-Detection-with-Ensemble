```python
# Import the modules
import matplotlib.pyplot as plot
import pandas as pd
# Initialize a dictionary for months
data = dict()
datasources = {'cohen':0, 'actmed':1, 'sirm':2, 'ricord':3, 'fig1':4, 'rsna':5, 'stonybrook': '6', 'bimcv':7}

data["Positive"] = [0]* len(datasources.keys())
data["Negative"] = [0]* len(datasources.keys())

with open('../input/covidx-cxr2/train.txt', 'r') as f:
    for line in f.readlines():
        # Store each line in the dictionary
        row = line.split(' ')
        label = row[2]
        datasource = row[3].strip()
        if(label == 'positive'):
            data['Positive'][int(datasources[datasource])] += 1
        else:
            data['Negative'][datasources[datasource]] += 1
            # Dictionary loaded into a DataFrame       
index = ["cohen", "actmed", "sirm", "ricord", "fig1", "rsna", "stonybrook", "bimcv"];
dataFrame = pd.DataFrame(data=data, index=index);
# Draw a vertical bar chart
dataFrame.plot.bar(rot=15, title="Positive vs Negative for various data sources");
print(dataFrame)
plot.show(block=True);
```
